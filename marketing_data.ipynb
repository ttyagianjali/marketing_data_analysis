```python
df = pd.read_csv('marketing_data.csv')
```

```python
df = pd.get_dummies(df, columns=['education','marital_status','country'], drop_first=True)
```

```python
df.columns = (
    df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

Y = df['sales_website']

X = df[['income','sales_deals', 'sales_catalog',
       'sales_store', 'website_visits','minor_children', 'age', 'total_spend',
       'education_basic','country_germany']]

X = np.asarray(X, dtype=float)
Y = np.asarray(Y, dtype=float)

X = sm.add_constant(X)

model = sm.OLS(Y, X).fit()

print(f'{model.summary()}')

 
```
